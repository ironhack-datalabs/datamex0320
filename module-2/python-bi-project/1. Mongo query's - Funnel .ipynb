## Funnel - Initial Query

```python
#1 seattle and surrondings
seattleq = list(colec.find({
                        '$and':[
                               {'$or':[{'offices.city': 'Seattle'},\
                               {'offices.city': 'Belleuve'},\
                               {'offices.city': 'Redmond'},\
                               {'offices.state_code': 'WA'}]},\
                        {'number_of_employees':{'$gt':5}},\
                        {'deadpooled_year': None }]},\
                        {'name':1, 'offices.city':1, '_id':0, 'founded_year':1,'number_of_employees':1,\
                         'offices.latitude':1, 'offices.longitude':1}))

```

```python
#2 Dataframe result as expected 
dfs=pd.DataFrame(list(seattleq))
dfs
```

```python
# 3 Assigning the correct column names to the unwrapped offices column
new_df = pd.DataFrame(index=[], columns=dfs.columns)
for _, i in dfs.iterrows():
    flattened_d = [dict(i.to_dict(), offices=c) for c in i.offices]
    new_df = new_df.append(flattened_d )
```

```python
#4 Fixing index
new_df = new_df.reset_index(drop=True) 
```

```python
#6 Fixing columns wrapped information
new_new = json_normalize(new_df['offices']) 
```

```python
#7 Merging DataFrames
df=pd.concat([new_df, new_new], axis=1).drop(columns=['offices'])
```

```python

df=pd.concat([new_df, new_new], axis=1).drop(columns=['offices'])
```

```python
#9 DataFrame as expected, cleaning process
df.dropna(inplace=True)
```

```python
df
```

```python
#10 Looking for the three main cities for the project
#data=df[(df["city"]== "Redmond")]
data = df.loc[(df['city'] == 'Seattle') | (df['city'] == 'Bellevue') | (df['city'] == 'Redmond')]
data
```

```python
#11 Reseting index
data.reset_index(drop=True, inplace=True)
```

```python
#12 Saving  dataset
data.to_csv('dataset_clean.csv')
```

## MAPS with Folium

```python
import folium
from folium import plugins
```

```python
#Reading clean dataset
df=pd.read_csv("dataset_clean.csv")
```

### Clustering with a Heat Map  in Seattle Great Area 

```python
#Generating Heat Map
mapa = folium.Map([47.616415, -122.253495], zoom_start=12)
value_data = data[['latitude', 'longitude']].values
mapa.add_child(plugins.HeatMap(value_data, radius=15))
mapa
```

```python
#Saving map
mapa.save('heatmap.html')
```

### Detail map of the Clusters in Seattle Great Area 

```python
#Generating markers map step 1
locations = df[['latitude', 'longitude']]
locationlist = locations.values.tolist()
len(locationlist)
locationlist[7]
```

```python
#Generating markers map step 2

map = folium.Map(location=[47.616415, -122.253495], zoom_start=12)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=df['name'][point]).add_to(map)
map
```

```python
#Saving map
map.save('detail_clustering.html')
```
