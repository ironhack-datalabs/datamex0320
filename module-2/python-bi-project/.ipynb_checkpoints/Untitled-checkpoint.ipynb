```python
import pymongo
import numpy as np
import pandas as pd
from pymongo import MongoClient
from pandas.io.json import json_normalize
import folium
import folium
from folium import plugins
```

```python
cliente=pymongo.MongoClient('mongodb://localhost:27017')
db=cliente.Companies
```

```python
paris = db.Companies.find({'$and':[{'offices.city': 'Paris'},{'category_code': 'games_video'},{'offices.latitude':{'$gte':1}},{'offices.longitude':{'$gte':1}}]},\
                    {'name':1, 'offices.city':1, '_id':0, 'offices.latitude':1, 'offices.longitude':1})
```

```python
paris
```

```python
df1=pd.DataFrame(paris)
df1
```

```python
df1=df1.dropna()
```

```python
dfll=pd.DataFrame()
for i in range(13):
    dfll=dfll.append(json_normalize(df1.offices[i]), ignore_index=True)

    
dfll
```

```python
df=pd.concat([df1, dfll], axis=1).drop(columns=['offices'])
```

```python
df
```

```python
df=df.drop([4,5,6,7,8,13,14,15,16,17,18])
```

```python
df.reset_index(drop=True)
```

```python
df.to_json('paris_gaming.json') 
```

```python
db.df.insert_many(df.to_dict('records'))
```
