```python
from tqdm import tqdm
```

```python
lista_df=[] #ver de utilizar el coso de barras 

for i in tqdm(range(len(data.offices))):
    test1 = pd.DataFrame(data.iloc[i].offices)
    test1['name'] = str(data.loc[i,['name'][0]])
    test1['total_employees'] = data.iloc[i].number_of_employees
    test1['category'] = str(data.loc[i,['category_code'][0]])
    test1['total_offices'] = len(data.iloc[i].offices)
    lista_df.append(test1)
```

```python
final = pd.concat(lista_df).reset_index(drop=True)
final.head()
```
