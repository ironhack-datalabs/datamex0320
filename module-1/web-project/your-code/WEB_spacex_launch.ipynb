```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import re
```

```python
url='https://en.wikipedia.org/wiki/List_of_Falcon_9_and_Falcon_Heavy_launches'
```

```python
html=requests.get(url).content 
soup=bs(html, 'html.parser')
```

```python
table2019=soup.find_all('table', {'class':'wikitable plainrowheaders'})[-2]
```

```python
rows=table2019.find_all('tr')
rows=[r.text.strip().split('\n') for r in rows]

final=[]
for e in rows:
    tmp=[]
    for st in e:
        if st!='':
            tmp.append(st)
    if len(tmp) > 3:
        final.append(tmp)
final[:3]
```

```python
#Different rows
rows1 = final[1:] 
Flight_number = [x[0] for x in rows1]
Date_and_time_UTC = [x[1] for x in rows1]
Version_Booster = [x[2] for x in rows1]
Launchsite = [x[3] for x in rows1]
Payload = [x[4] for x in rows1]
Payload_mass = [x[5] for x in rows1]
Orbit = [x[6] for x in rows1]
Customer = [x[7] for x in rows1]
Launch_outcome = [x[8] for x in rows1]
Booster_Landig = [x[9] for x in rows1]
```

```python
df_ls = pd.DataFrame({'Flight Number':Flight_number,
                      'Date':Date_and_time_UTC,
                      'Time (UTC)':Date_and_time_UTC,
                      'Booster Version':Version_Booster,
                      'Launch Site':Launchsite,
                      'Payload':Payload,
                      'Payload Mass (kg)':Payload_mass,
                      'Orbit':Orbit,
                      'Customer':Customer,
                      'Mission Outcome':Launch_outcome,
                      'Landing Outcome':Booster_Landig,})
```

```python
df_ls
```

```python
#Function to filter the date with the format year-month-day
def replace_d(x):
    y = str(x)
    match = re.search('((.*?) .+, (.){4})', y)
    b = match
    c = 0
    if 'January' in str(b):
        c = str(b[0]).replace('January', '01')
    elif 'February' in str(b):
        c = str(b[0]).replace('February', '02')
    elif 'March' in str(b):
        c = str(b[0]).replace('March', '03')
    elif 'April' in str(b):
        c = str(b[0]).replace('April', '04')
    elif 'May' in str(b):
        c = str(b[0]).replace('May', '05')
    elif 'June' in str(b):
        c = str(b[0]).replace('June', '06')
    elif 'July' in str(b):
        c = str(b[0]).replace('July', '07')
    elif 'August' in str(b):
        c = str(b[0]).replace('August', '08')
    elif 'November' in str(b):
        c = str(b[0]).replace('November', '11')
    elif 'December' in str(b):
        c = str(b[0]).replace('December', '12')
    d = str(c).replace(',','')
    e = str(d).split(' ')
    if len(e[1]) < 2:
        e[1] = ('0{}'.format(e[1]))
    f = str('{}-{}-{}'.format(e[2],e[0],e[1]))    
    return f
df_ls['Date'] = df_ls['Date'].apply(replace_d) 
```

```python
#Function to filter the hour
def replace_h(x):
    y = str(x)
    match = re.search('.{2}:.{2}',y)
    return ('{}:00'.format(str(match[0])))

df_ls['Time (UTC)'] = df_ls['Time (UTC)'].apply(replace_h) 
```

```python
#Function to replace '♺' in Booster version
df_ls['Booster Version'] = df_ls['Booster Version'].apply(lambda x: x.replace('♺','')) 
```

```python
#Function to eliminate what is in ()
df_ls['Payload'] = df_ls['Payload'].apply(lambda x: (x.split('('))[0])
```

```python
#Function to eliminate mass in lb placed into '()'
df_ls['Payload Mass (kg)'] = df_ls['Payload Mass (kg)'].apply(lambda x: (x.split('('))[0])
```

```python
#Function to get rid of the 'kg' word
df_ls['Payload Mass (kg)'] = df_ls['Payload Mass (kg)'].apply(lambda x: x.replace('kg',''))
```

```python
#Function to eliminate ()
df_ls['Landing Outcome'] = df_ls['Landing Outcome'].apply(lambda x: x.split('(')[0] if '(' in str(x) else x)
```
