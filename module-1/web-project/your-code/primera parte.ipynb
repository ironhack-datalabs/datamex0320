```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup as bs
```

```python
#importo el csv que contiene lyrics de diferentes canciones
dforiginal=pd.read_csv('songdata.csv')
```

```python
#Exploro el dataset para saber qué columnas contiene
dforiginal.columns
```

```python
#Exploración para saber qué artistas contiene
dforiginal.artist.unique()
```

```python
#Selección de las filas que contienen como autora a Lady Gaga
quiero=list(dforiginal.artist[dforiginal.artist=='Lady Gaga'].index)
```

```python
len(quiero)
```

```python
#Del dataset original, localizo las filas que contienen canciones de Lady gaga 
#Selecciono únicamente esas filas.
df_gaga=dforiginal.iloc[quiero][['artist', 'song', 'text']]
```

```python
df_gaga2=df_gaga.copy()
```

```python
df_gaga2.song.unique
```

```python
#Reseteo el índice y obtengo un nuevo dataframe 
df_gaga2.reset_index()
```

```python
df_gaga2=df_gaga2.drop('artist',axis=1)
```

```python
df_gaga2.song.unique()
```

```python
#Scraping para obtener las canciones del album The Fame 
url='https://ladygaga.fandom.com/es/wiki/The_Fame_(%C3%A1lbum)'
res=requests.get(url)
data=res.text
soup=bs(data, 'html.parser')
lista=soup.find_all('ol')
lista=lista[3].find_all('li')
thefame=[]
for i in range(17):
    nombre=lista[i].find('a')
    thefame.append(nombre.text)
thefame
```

```python
#conditions = [(df_gaga2['song'] in thefame==True)]
#df_gaga2['Album'] = np.select(conditions,[(df_gaga2['Album']=='The Fame')], default='Single')
```

```python
#Scraping para obtención de cancioens del álbum Born this way.
url2='https://ladygaga.fandom.com/es/wiki/Born_This_Way_(%C3%A1lbum)#Est%C3%A1ndar'
res=requests.get(url2)
data=res.text
soup=bs(data, 'html.parser')
lista=soup.find_all('ol')
lista=lista[1].find_all('li')
borntw=[]
for i in range(14):
    nombre=lista[i].find('a')
    borntw.append(nombre.text)
borntw
```

```python
#Obtención de canciones del disco de The Fame Monster
url3='https://wikiladygaga.fandom.com/es/wiki/The_Fame_Monster_(%C3%A1lbum)'
res=requests.get(url3)
data=res.text
soup=bs(data, 'html.parser')
lista=soup.find_all('ol')
lista=lista[1].find_all('li')
monster=[]
for i in range(8):
    nombre=lista[i].find('a')
    monster.append(nombre.text)
monster
```

```python
#Obtención de canciones del album de Artpop
url3='https://ladygaga.fandom.com/wiki/Artpop_(album)'
res=requests.get(url3)
data=res.text
soup=bs(data, 'html.parser')
lista=soup.find_all('table')
artpop=[]
lista=lista[1].find_all('tr')
for i in range(15):
    lista2=lista[i].find_all('td')
    lista2=lista2[1].find('a')
    artpop.append(lista2.text)
artpop[7]='Artpop'
artpop
```

```python
#Obtención de canciones del álbum Joanne
url4='https://ladygaga.fandom.com/wiki/Joanne_(album)'
res=requests.get(url4)
data=res.text
soup=bs(data, 'html.parser')
lista=soup.find_all('table')
joanne=[]
lista=lista[4].find_all('tr')
for i in range(15):
    lista2=lista[i].find_all('td')
    lista2=lista2[1].find('a')
    joanne.append(lista2.text)
#lista2=lista[14].find_all('td')
#lista2=lista2[1].find('a')
joanne
```

```python
#Creación de columna de Album
#De acuerdo con la canción se clasifica por álbum
col_album=[]
for e in df_gaga2['song']:
    if e in thefame:
        col_album.append('The Fame')
    elif e in borntw:
        col_album.append('Born this Way')
    elif e in monster:
        col_album.append('The Fame Monster')
    elif e in artpop:
        col_album.append('Artpop')
    elif e in joanne:
        col_album.append('Joanne')
    else:
        col_album.append('Single')
col_album
```

```python
#Se anexa la columna de álbum al dataframe y posteriormente se resetea
#el índice  
df_gaga2['Album']=col_album
```

```python
df_gaga2=df_gaga2.reset_index()
```

```python
df_gaga2=df_gaga2.drop('index',axis=1)
df_gaga2
```

```python
df_gaga2.to_csv('LadyGaga1.csv') #Almaceno en un csv como copia
```
