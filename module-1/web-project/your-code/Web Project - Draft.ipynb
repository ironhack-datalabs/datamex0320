```python
def tabla(tabla):
    
    #extract rows from soup
    rows=tabla.find_all('tr')
    rows=[r.text.strip().split('\n') for r in rows]
    tabla0 = []
    for e in rows:
        tmp=[]
        for st in e:
            if st!='':
                tmp.append(st)
        tabla0.append(tmp)
        
    
    #create df
    col_names = tabla0[0]
    data = tabla0[1:]
    df = pd.DataFrame(data,columns=col_names)
    
    #get the bad indexes in year column
    bad_index = list(df.Year[df.Year.str.isnumeric() == False].index)
    
    
    #clean bad indexes
    for i in bad_index:
        df.iloc[i] = df.iloc[i].shift(periods=1)
    
    
    #fill values in year column
    df.Year = df.Year.ffill(axis=0)
    
    
    #clean the game column
    df.Game = df.Game.apply(lambda x : re.sub('\[\d+\]','', x))
    

    
    return df
```

```python
def totalsales(nombre):
    try:
        nombre = str(nombre).replace(' ','+')    
        url2 = 'https://www.vgchartz.com/games/games.php?name={}&showtotalsales=1'.format(nombre)
        respuesta = req.get(url2).content
        soup = bs(respuesta,'lxml')
        general_body = soup.find_all('div',{'id':'generalBody'})
        rows=general_body[0].find_all('tr')
        rows=[r.text.strip().split('\n') for r in rows]
        final=[]
        for e in rows:
            tmp=[]
            for st in e:
                if st!='':
                    tmp.append(unidecode.unidecode(st))
            final.append(tmp)

        col_names=final[2]
        data=final[3:-1]

        df=pd.DataFrame(data, columns=col_names)
        df['Total Sales'] = df['Total Sales'].apply(lambda x : x.split('m')[0])

        df['Total Sales'] = pd.to_numeric(df['Total Sales'], errors='coerce')

        df['Total Sales'] = df['Total Sales'].astype(float)

        x = df['Total Sales'].sum()

        return x
    
    except:
        return 'Unknown'
```

```python
from tqdm import tqdm, trange
from random import random, randint
from time import sleep

with trange(10) as t:
    for i in t:
        # Description will be displayed on the left
        t.set_description('GEN %i' % i)
        # Postfix will be displayed on the right,
        # formatted automatically based on argument's datatype
        t.set_postfix(loss=random(), gen=randint(1,999), str='h',
                      lst=[1, 2])
        sleep(0.1)

with tqdm(total=10, bar_format="{postfix[0]} {postfix[1][value]:>8.2g}",
          postfix=["Batch", dict(value=0)]) as t:
    for i in range(10):
        sleep(0.1)
        t.postfix[1]["value"] = i / 2
        t.update()
```
