```python
def tabla(tabla):
    
    #extract rows from soup
    rows=tabla.find_all('tr')
    rows=[r.text.strip().split('\n') for r in rows]
    tabla0 = []
    for e in rows:
        tmp=[]
        for st in e:
            if st!='':
                tmp.append(st)
        tabla0.append(tmp)
        
    
    #create df
    col_names = tabla0[0]
    data = tabla0[1:]
    df = pd.DataFrame(data,columns=col_names)
    
    #get the bad indexes in year column
    bad_index = list(df.Year[df.Year.str.isnumeric() == False].index)
    
    
    #clean bad indexes
    for i in bad_index:
        df.iloc[i] = df.iloc[i].shift(periods=1)
    
    
    #fill values in year column
    df.Year = df.Year.ffill(axis=0)
    
    
    #clean the game column
    df.Game = df.Game.apply(lambda x : re.sub('\[\d+\]','', x))
    

    
    return df
```

## 3. Completing CSV with API

```python
#api: https://api.rawg.io/api/games
```

```python
%%time
import time
from tqdm import tqdm

#defining lists to get a DF
lista_name = []
lista_platform = []
lista_year = []
lista_genre = []

url = 'https://api.rawg.io/api/games?dates=2017-01-01%2C2020-04-10&page_size=40&page='
pagina=1

#loop to get 4000 rows. Every page has 40 results. Using TQDM to get the progress bar
with tqdm(total=100) as pbar:
    while pagina<=100:
        for i in range(40):
            resp=req.get(url+str(pagina))
            api = resp.json()
            #getting name column
            lista_name.append(api['results'][i]['name'])
            #getting platform column
            if api['results'][i]['platforms']:
                for e in range(len(api['results'][i]['platforms'])):
                    plataforma = []
                    plataforma.append(api['results'][i]['platforms'][e]['platform']['name'])
            else:
                plataforma=[]
            lista_platform.append(plataforma)
            #getting year column. Comes as a full date, we have to clean it after.
            lista_year.append(api['results'][i]['released'])
            #getting the genre column
            if api['results'][i]['genres']:
                for a in range(len(api['results'][i]['genres'])):
                    genero = []
                    genero.append(api['results'][i]['genres'][a]['name'])
            else:
                genero=[]
            lista_genre.append(genero)
        pagina +=1
        pbar.update(1)
```

```python
#constructing DF with the information we got in last step
listas_api = pd.DataFrame(
    {'Name': lista_name,
     'Platform': lista_platform,
     'Year_of_Release': lista_year,
     'Genre': lista_genre,
    })
```

```python
listas_api.head()
```

```python
#cleaning the Year_of_Release column. The year is kept. 
listas_api['Year_of_Release'] = listas_api['Year_of_Release'].apply(lambda x: x.split('-')[0])
```

```python
listas_api
```

```python
def totalsales(nombre):
    '''This function gets the total sales in vgchartz.com based on the name. It searches the name and sums the results'''
    
    try:
        nombre = str(nombre).replace(' ','+')    
        url2 = 'https://www.vgchartz.com/games/games.php?name={}&showtotalsales=1'.format(nombre)
        respuesta = req.get(url2).content
        soup = bs(respuesta,'lxml')
        general_body = soup.find_all('div',{'id':'generalBody'})
        rows=general_body[0].find_all('tr')
        rows=[r.text.strip().split('\n') for r in rows]
        final=[]
        for e in rows:
            tmp=[]
            for st in e:
                if st!='':
                    tmp.append(unidecode.unidecode(st))
            final.append(tmp)

        col_names=final[2]
        data=final[3:-1]

        df=pd.DataFrame(data, columns=col_names)
        df['Total Sales'] = df['Total Sales'].apply(lambda x : x.split('m')[0])

        df['Total Sales'] = pd.to_numeric(df['Total Sales'], errors='coerce')

        df['Total Sales'] = df['Total Sales'].astype(float)

        x = df['Total Sales'].sum()

        return float(x)
    
    except:
        return np.nan
```

```python
%%time
#applying last function to dataframe. Using TQDM to show the progressbar
tqdm.pandas()
listas_api['Global_Sales'] = listas_api['Name'].progress_apply(totalsales)
```

```python
listas_api
```

```python
#Concatenating the dataframe got from the CSV and the one got from the API
subfinaldf = pd.concat([df, listas_api], ignore_index=True)
```

```python
#Grouping by sales. 
#Some games were counting by platform. So in case there are many platforms, the result is the sum of all of them

dfsales = subfinaldf.groupby('Name')['Global_Sales'].sum()
dfsales[:5]
```

# Merging everything

```python
#Merging Global Sales column to finaldf
finaldf = finaldf.merge(dfsales, on = 'Name', how = 'left')
finaldf.head()
```
