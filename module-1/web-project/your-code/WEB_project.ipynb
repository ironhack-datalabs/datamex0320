```python
#Replacing all the data from API 2019 to more complete data taken from WEB scraping wikipedia
data_replace = data_original_api.copy()
data_replace.loc[73:85,'Date':'Landing Outcome']=data_original_web.loc[0:13,'Date':'Landing Outcome'].values
data_replace[73:80]
```

```python
data_replace.loc[5:61,'Date':'Landing Outcome']=data_original_csv.loc[0:56,'Date':'Landing Outcome'].values
```

```python
data_replace.head(10)
```

```python
data_replace.fillna('Unk')
data_replace.head(20)
```

```python
#Cleaning time(UTC) column
stage_cleaning = data_replace.copy() 
def adding_zero(x):
    y = x.split(':')
    if len(y[0])<2:
        return ('0{}:{}:{}'.format(y[0],y[1],y[2]))
    else:
        return x
stage_cleaning['Time (UTC)'] = stage_cleaning['Time (UTC)'].apply(adding_zero) 
```

```python
#Getting rid of the Falcon word in column Booster Version
def Falcon(x):
    if 'Falcon 1' in x:
        y = x.split(' ')
        return ('{}{} v1.0'.format(y[0].replace('Falcon','F'),y[1]))
    elif 'Falcon 9' in x:
        y = x.split(' ')
        return ('{}{} v1.1'.format(y[0].replace('Falcon','F'),y[1]))
    else:
        return x
stage_cleaning['Booster Version'] = stage_cleaning['Booster Version'].apply(Falcon)
stage_cleaning['Booster Version'] = stage_cleaning['Booster Version'].apply(lambda x: (x.split('['))[0] if '[' in str(x) else x)
stage_cleaning['Booster Version'] = stage_cleaning['Booster Version'].apply(lambda x: x.replace('  ',' ') if '  ' in str(x) else x)
```

```python
#Eliminating character 'Â' from the mass
stage_cleaning['Payload Mass (kg)'] = stage_cleaning['Payload Mass (kg)'].apply(lambda x: str(x).replace('Â','') if 'Â' in str(x) else x)
```

```python
#Landing outcome cleaning
stage_cleaning1 = stage_cleaning.fillna('Unk')
stage_cleaning1['Landing Outcome'] = stage_cleaning1['Landing Outcome'].apply(lambda x: (x.split('('))[0] if '(' in str(x) else x)
stage_cleaning1['Landing Outcome'] = stage_cleaning1['Landing Outcome'].apply(lambda x: (x.split('['))[0] if '[' in str(x) else x)
```

```python
#We don´t want [...] in the column Payload
stage_cleaning1['Payload'] = stage_cleaning1['Payload'].apply(lambda x: (x.split('['))[0] if '[' in str(x) else x)
```

```python
#Getting rid of quotations '' in Customer column
stage_cleaning1['Customer'] = stage_cleaning1['Customer'].apply(lambda x: (x.split("'"))[1] if "'" in str(x) else x)
```

```python
#We only want a maximum of 3 words separated by spaces
def payload_s(x):
    y = str(x).split(' ')
    if len(y) > 3:
        return ('{} {} {}'.format(y[0],y[1],y[2]))
    else:
        return x
stage_cleaning1['Payload'] = stage_cleaning1['Payload'].apply(payload_s)
```
