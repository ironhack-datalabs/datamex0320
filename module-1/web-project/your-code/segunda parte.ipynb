```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup as bs
```

```python
df=pd.read_csv('LadyGaga1.csv')
df=df.drop('Unnamed: 0',axis=1)
```

```python
#Se carga el último dataframe construido y se separan entre las 
#canciones que pertenecen a un álbum 
df_album=df[(df['Album']!='Single')]
df_album
```

```python
#Scraping para obtener autor y duración de las canciones de los álbumes

urlfame='https://es.wikipedia.org/wiki/The_Fame'
html=requests.get(urlfame).content 
soup=bs(html, 'lxml')
table=soup.find_all('table', {'class':'tracklist'})
autor=[]
song=[]
dur=[]
for i in range(2,15):
    final=table[1].find_all('tr')
    final=final[i].find_all('td')
    song.append(final[1].text)
    autor.append(final[2].text)
    dur.append(final[3].text)
dic={'song':song,'autor':autor,'duración':dur}

fame=pd.DataFrame(dic)
fame
```

```python
songs=[]
for e in fame['song']:
    aux=(re.findall('[a-zA-Z]+',e))
    songs.append(' '.join(aux))
songs[0]='Just Dance'
songs[5]='Beautiful, Dirty, Rich'
songs[4]='Eh, Eh (Nothing Else I Can Say)'
songs[8]='Starstruck'
songs
```

```python
fame['song']=songs
```

```python
fame
```

```python
urlfamem='https://es.wikipedia.org/wiki/The_Fame_Monster'
html=requests.get(urlfamem).content 
soup=bs(html, 'lxml')
table=soup.find_all('table', {'class':'tracklist'})
autor=[]
song=[]
dur=[]

for i in range(2,9):
    final=table[2].find_all('tr')
    final=final[i].find_all('td')
    song.append(final[1].text)
    autor.append(final[2].text)
    dur.append(final[4].text)
dic={'song':song,'autor':autor,'duración':dur}

famem=pd.DataFrame(dic)
famem
```

```python
#Función para limpiar las strings de los nombres de las canciones
def limpi_song(x):
    songs=[]
    for e in x:
        aux=(re.findall('[a-zA-Z]+',e))
        songs.append(' '.join(aux))
    return songs
```

```python
url='https://es.wikipedia.org/wiki/Joanne_(%C3%A1lbum)'
html=requests.get(url).content 
soup=bs(html, 'lxml')
table=soup.find_all('table', {'class':'tracklist'})
autor=[]
song=[]
dur=[]

for i in range(2,13):
    final=table[0].find_all('tr')
    final=final[i].find_all('td') #Para variar fila de escritor (2-12)
    song.append(final[1].text) #para variar columna 
    autor.append(final[2].text)
    dur.append(final[3].text)

dic={'song':song,'autor':autor,'duración':dur}
joanne=pd.DataFrame(dic)
joanne
```

```python
#scraping año de lanzamiento
url='https://ladygaga.fandom.com/wiki/Discography'
html=requests.get(url).content 
soup=bs(html, 'lxml')
tabla=soup.find_all('div', {'class':'supatext'})
albumaño=[]
año=[]
for i in range(6):
    albumaño.append(tabla[i].text)
albumaño.pop(4)
albumaño
import re
for e in albumaño:
    año.append(re.findall('[0-9]+',e))
año
```

```python
#CSV obtenido a partir de la API de lastfm
lastfm=pd.read_csv('lastfmapi.csv')
lastfm[['name','url','playcount']]
```

```python
#Cuenta de filas que contienen los álbumes The Fame
fame=list(lastfm.name[(lastfm.name.str.startswith('The Fame'))].index)
fame
```

```python
#Cuenta el número de reproducciones de dicho álbum
sumafame=lastfm.iloc[fame][['playcount']].sum()
sumafame
```

```python
born=list(lastfm.name[(lastfm.name.str.startswith('Born'))].index)
sumaborn=lastfm.iloc[born][['playcount']].sum()
art=list(lastfm.name[(lastfm.name.str.startswith('ART'))].index)
sumaart=lastfm.iloc[art][['playcount']].sum()
joa=list(lastfm.name[(lastfm.name.str.startswith('JO'))|(lastfm.name.str.startswith('Jo'))].index)
sumajoanne=lastfm.iloc[joa][['playcount']].sum()
```

```python
col_plays=[]
for e in dataGaga['Album']:
    if e=='The Fame':
        col_plays.append(sumafame[0])
    elif e=='The Fame Monster':
        col_plays.append(sumafame[0])
    elif e=='Born this Way':
        col_plays.append(sumaborn[0])
    elif e=='Artpop':
        col_plays.append(sumaart[0])
    else:
        col_plays.append(sumajoanne[0])
```

```python
dataGaga['Reproducciones']=col_plays
```
