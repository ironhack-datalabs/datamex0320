```python
# Libraries
import pandas as pd
import json
from pandas.io.json import json_normalize
import requests
```

```python
Api_rick = requests.get('https://rickandmortyapi.com/api/')
Api_rick.json()
```

```python
#Data frame 
flattened_data = json_normalize(Api_rick.json())

flattened_data.head()
```

```python
#Data characters
Api_char = requests.get('https://rickandmortyapi.com/api/character')
flattened_data2 = json_normalize(Api_char.json()['results'])
flattened_data2
```
